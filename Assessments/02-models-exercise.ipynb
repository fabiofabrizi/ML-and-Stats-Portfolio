## Plotting the Absolute Value Function

As has been the pattern till now, generate some fake data to use in the plot of the absolute value function.<br>
The general form of the absolute value function can be expressed as:

*f(x) = a|x - h| + k*<br>

where *a* represents how far the graph stretches vertically, *h* represents the horizontal shift and *k* represents the vertical shift.
If *a* is negative the graph opens downwards and if it's positive it opens upwards

First, we can start with a small numpy array randomly generated:

```python
# Generate an array of 60 elements between -50 and 50
array5 = np.linspace(start = -50, stop = 50, num = 60)
array6 = np.absolute(array5)                 
print(array6)
# Plot x vs y
plt.plot(array5, np.absolute(array5), label='y=|x|')
plt.legend(loc='upper center')
#plt.plot(array5)
plt.show()
```

The example below is adding more randomness into the variables.

```python
# Generate a random number for m. These are the number of data points.
# If the number drops too low, then it makes it harder to see the function.
# Also, each time the cell is run, a different plot is displayed.
m = random.randrange(50, 1000)
n = 10
X = 180 * np.random.uniform(size=(m, n)) - 90
y = np.abs(X[:, 8] - 4.0) + 1 * np.random.normal(size=m)

# Plot the model
plt.figure()
plt.plot(X[:, 8], y, 'b.')
plt.show()
```

## Absolute Value Function:
### Why it's not typically used to fit straight lines to data

If we consider the first plot below, which is just the same one as above, but with the plot of 
the array (x):

```python
# Generate an array of 60 elements between -50 and 50
array5 = np.linspace(start = -50, stop = 50, num = 60)
array6 = np.absolute(array5) 
print(array5)
print("\n")
print(array6)
# Plot x vs y for absolute value function
#plt.plot(array5, np.absolute(array5))
# Plot y = x

plt.plot(array5, 'g-', label=r"$x$")
plt.plot(array6, 'b.', label=r"$|x|$")

# Add a legend.
plt.legend()
plt.show()
```

So we see that when the Absolute Value Function is used to fit a straight line to the data, we lose all the information that the negative values give us. 

A linear function is of the form:<br>

*y = mx + c*<br>

Which is a straight line, whereas the Absolute Value Function has a sharp edge as a turning point: at *abs(x)* = 0<br>
<br>
And so, because it's not a linear function, it's not the best tool to fit lines to data, especially if the data has plenty of negative values.

**More Examples**<br>
consider the following:<br>

*y = 2x - 5*<br>
*y = -2(x - 6) + 9*

And their absolute value functions:<br>

*f(x) = |2x - 5|*<br>
*g(x) = -2|(x - 6)| + 9*

Let's generate some random data for x:

```python
# Give values from -10 to 21, in steps of 1.0
x_data = np.arange(-10, 21, 1)
x_data
# y = 2x - 5
y_data = (2 * x_data) - 5
# absolute value function: |2x - 5|
abs_y_data = np.absolute(y_data)

# Print the arrays
print(f'x_data is:\n {x_data}\n')
print(f'y_data is:\n {y_data}')

# Plot the data
# y = 2x - 5
plt.plot(x_data, y_data, 'g-', label=r"$2x - 5$")

plt.plot(x_data, abs_y_data, 'r.', label=r"$|2x-5|$")

# Display labels and plots
plt.legend()
plt.show()
```

**The second example**<br>
We'll use the same data for *x* that was generated above:

*y = -2(x - 6) + 9*
<br>
and<br>
*g(x) = -2|(x - 6)| + 9*

```python
x_data
y_data2 = -2 * (x_data - 6) + 9
# absolute value function: |2x - 5|
x_abs = np.absolute(x_data - 6)
abs_y_data2 = (-2 * x_abs) + 9

# Print x and y
print(f'x_data is:\n {x_data}\n')
print(f'y_data is:\n {y_data2}')
print(f'x_abs is:\n{x_abs}')

# Plot the data
# y = -2(x - 6) + 9
plt.plot(x_data, y_data, 'g-', label=r"$-2(x - 6) + 9$")
# y = -2|(x - 6)| + 9
plt.plot(x_data, abs_y_data2)
```

### Conclusion

The absolute value function can only be used to fit straight lines up until the 'sharp' turning point. We've seen above how with a negative value of *a*, the graph opens downwards.<br>

Linear functions are straight lines.
As *abs(x)* has a 'sharp edge' it can't be classed as linear and therefore would not be the best tool to use for fitting straight lines.
