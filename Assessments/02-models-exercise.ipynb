## Plotting the Absolute Value Function

As has been the pattern till now, generate some fake data to use in the plot of the absolute value function.<br>
The general form of the absolute value function can be expressed as:

*f(x) = a|x - h| + k*<br>

where *a* represents how far the graph stretches vertically, *h* represents the horizontal shift and *k* represents the vertical shift.
If *a* is negative the graph opens downwards and if it's positive it opens upwards

First, we can start with a small numpy array randomly generated:

```python
# Generate an array of 60 elements between -50 and 50
array5 = np.linspace(start = -50, stop = 50, num = 60)
array6 = np.absolute(array5)                 
print(array6)
# Plot x vs y
plt.plot(array5, np.absolute(array5), label='y=|x|')
plt.legend(loc='upper center')
#plt.plot(array5)
plt.show()
```

The example below is adding more randomness into the variables.

```python
# Generate a random number for m. These are the number of data points.
# If the number drops too low, then it makes it harder to see the function.
# Also, each time the cell is run, a different plot is displayed.
m = random.randrange(50, 1000)
n = 10
X = 180 * np.random.uniform(size=(m, n)) - 90
y = np.abs(X[:, 8] - 4.0) + 1 * np.random.normal(size=m)

# Plot the model
plt.figure()
plt.plot(X[:, 8], y, 'b.')
plt.show()
```

## Absolute Value Function:
### Why it's not typically used to fit straight lines to data

If we consider the first plot below, which is just the same one as above, but with the plot of 
the array (x):

```python
# Generate an array of 60 elements between -50 and 50
array5 = np.linspace(start = -50, stop = 50, num = 60)
array6 = np.absolute(array5) 
print(array5)
print("\n")
print(array6)
# Plot x vs y for absolute value function
#plt.plot(array5, np.absolute(array5))
# Plot y = x

plt.plot(array5, 'g-', label=r"$x$")
plt.plot(array6, 'b.', label=r"$|x|$")

# Add a legend.
plt.legend()
plt.show()
```

So we see that when the Absolute Value Function is used to fit a straight line to the data, we lose all the information that the negative values give us. 

A linear function is of the form:<br>

*y = mx + c*<br>

Which is a straight line, whereas the Absolute Value Function has a sharp edge as a turning point: at *abs(x)* = 0<br>
<br>
And so, because it's not a linear function, it's not the best tool to fit lines to data, especially if the data has plenty of negative values.

**More Examples**<br>
consider the following:<br>

*y = 2x - 5*<br>
*y = -2(x - 6) + 9*

And their absolute value functions:<br>

*f(x) = |2x - 5|*<br>
*g(x) = -2|(x - 6)| + 9*

Let's generate some random data for x:

```python
# Give values from -10 to 21, in steps of 1.0
x_data = np.arange(-10, 21, 1)
x_data
# y = 2x - 5
y_data = (2 * x_data) - 5
# absolute value function: |2x - 5|
abs_y_data = np.absolute(y_data)

# Print the arrays
print(f'x_data is:\n {x_data}\n')
print(f'y_data is:\n {y_data}')

# Plot the data
# y = 2x - 5
plt.plot(x_data, y_data, 'g-', label=r"$2x - 5$")

plt.plot(x_data, abs_y_data, 'r.', label=r"$|2x-5|$")

# Display labels and plots
plt.legend()
plt.show()
```

**The second example**<br>
We'll use the same data for *x* that was generated above:

*y = -2(x - 6) + 9*
<br>
and<br>
*g(x) = -2|(x - 6)| + 9*

```python
x_data
y_data2 = -2 * (x_data - 6) + 9
# absolute value function: |2x - 5|
x_abs = np.absolute(x_data - 6)
abs_y_data2 = (-2 * x_abs) + 9

# Print x and y
print(f'x_data is:\n {x_data}\n')
print(f'y_data is:\n {y_data2}')
print(f'x_abs is:\n{x_abs}')

# Plot the data
# y = -2(x - 6) + 9
plt.plot(x_data, y_data, 'g-', label=r"$-2(x - 6) + 9$")
# y = -2|(x - 6)| + 9
plt.plot(x_data, abs_y_data2)
```

### Conclusion

The absolute value function can only be used to fit straight lines up until the 'sharp' turning point. We've seen above how with a negative value of *a*, the graph opens downwards.<br>

Linear functions are straight lines.
As *abs(x)* has a 'sharp edge' it can't be classed as linear and therefore would not be the best tool to use for fitting straight lines.

An explanation of the Cost Function (from Ian's notebook, see references)

The best line is the one that minimises the following calculated value.

$$ \sum_i (y_i - mx_i - c)^2 $$

Here $(x_i, y_i)$ is the $i^{th}$ point in the data set and $\sum_i$ means to sum over all points. 

The values of $m$ and $c$ are to be determined.


We usually denote the above as $Cost(m, c)$.
Where does the above calculation come from?

It's easy to explain the part in the brackets $(y_i - mx_i - c)$.

The corresponding value to $x_i$ in the dataset is $y_i$.

These are the measured values.

The value $m x_i + c$ is what the model says $y_i$ should have been.

The difference between  the value that was observed ($y_i$) and the value that the model gives ($m x_i + c$), is $y_i - mx_i - c$.

**We can use *Least Squares Fitting* together with numpy**
Note how the cost below is the python implementation of the equation above.

```python
# Calculate the cost of the lines above for the x, y data specified in the task
cost = lambda m,c: np.sum([(y[i] - m * x[i] - c)**2 for i in range(x.size)])

print("Cost with m = %5.2f and c = %5.2f: %8.2f" % (0.08, 0.75, cost(0.08, 0.75)))
print("Cost with m = %5.2f and c = %5.2f: %8.2f" % (0.11, 0.97, cost(0.11, 0.97)))
print("Cost with m = %5.2f and c = %5.2f: %8.2f" % (0.1, 0.97, cost(0.1, 0.97)))
```

Initially, this makes sense, as it can be seen from the graph above that the estimate used to plot the red line is way off.<br>
The other two lines have cost function values closer to each other and they also go through more data points then the red line.

Moving forward, the estimated values for the green line will be chosen in any calculations to optimise.

### Minimising the cost

We want to calculate what the lowest values of $m$ and $c$ will be.<br>

1) Calculate the means (averages) of x and y
2) Subtract the means from x and y

```python
# Calculate the best values for m and c.

# First calculate the means (a.k.a. averages) of x and y.
x_avg = np.mean(x)
y_avg = np.mean(y)

# Subtract means from x and y.
x_zero = x - x_avg
y_zero = y - y_avg

# The best m is found by the following calculation.
m = np.sum(x_zero * y_zero) / np.sum(x_zero * x_zero)
# Use m from above to calculate the best c.
c = y_avg - m * x_avg

print("m is %8.6f and c is %6.6f." % (m, c))
```

Now we're getting closer to finding the optimal line - These values are close to the green line from the estimate which fit most of the data points.<br>

We can now use [numpy polyfit](https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html), which is a method that fits the data within a polynomial function. 
It works by finding the best fitting curve to a given set of points (the data points for *x* and *y* in this case) by minimising the sum of squares.

In the equation below, 1 is the degree of the polynomial (straight line).

```python
np.polyfit(x, y, 1)
```

And we can see that we have the same numbers as above - so numpy polyfit has confirmed the hand calculations.

### Best fit line

Using least squares fitting we have approximate values of
$m$ = 0.1031
and 
$c$ = 1.0495

And we can now plot the best fit line on top of the original data below.

```python
# Plot the best fit line.
fig, ax = plt.subplots(figsize=(8, 6))
plt.plot(x, y, 'k.', label='Original data')
plt.plot(x, m * x + c, 'b-', label='Best fit line')

# Add axis labels and a legend.
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend()

# Show the plot.
plt.show()
```

### Optimisation

We can try to optimise the parameters further by using scipy
to estimate the parameters.


$$Cost(m, c) = \sum_i (y_i - mx_i - c)^2 $$

```python
# Optimization.
import scipy.optimize as so

# Numerical arrays.
import numpy as np
```

```python
y = 0.11 * x + 0.97 + np.random.normal(0.0, 0.11, x.size)

# print them out:
x, y
```

Now write a function to unpack m and c from the above results.

```python
def cost(MC):
    # Unpack m and c
    m, c = MC
    cost = np.sum((y - m * x - c)**2)
    # return cost
    return cost
```

```python
# run quick test:
cost((0.11, 0.97))
```

Now use [scipy optimise minimise](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html) to try and get further optimised value for m & c.

It's a procedure used in this case to find the minimum values of our function (m & c).
Inputs are the cost function and our estimates.

```python
result = so.minimize(cost, (0.11, 0.97))
# show the results
result
```

```python
# Extract the optimized m and c.
m_o, c_o = result.x
# Print them.
m_o, c_o
```

```python
# Previous, analytical result.
m_a, c_a = np.polyfit(x, y, 1)
# Print analytical results.
m_a, c_a
```

```python
# Plot the best fit line from the optimization.
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, 'k.', label='Original data')
ax.plot(x, m_o * x + c_o, 'b-', label='Optimized line')
ax.plot(x, m_a * x + c_a, 'g-', label='Analytical line')
ax.legend()

# Show the plot.
plt.show()
```

### Curve Fitting

For the curve fitting we use the [scipy.optimize.curve_fit](https://docs.scipy.org/doc/scipy/reference/generated/scipy.optimize.minimize.html) package. 

It's another scipy method that finds the best-fit parameters using least-squares fit.

```python
def f(x, m, c):
    return m * x + c
```

```python
# Run the curve fit
result = so.curve_fit(f, x, y)
```

```python
# Look at the result.
result
```

```python
# get the parameters for m & c from the array
m_f, c_f = result[0]
# Print values.
m_f, c_f
```

```python
# Plot the best fit line from the optimization.
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, y, 'k.', label='Original data')
ax.plot(x, m_f * x + c_f, 'r-', label='Curve Fit line')
ax.plot(x, m_a * x + c_a, 'g-', label='Analytical line')
ax.legend()

# Show the plot.
plt.show()
```
